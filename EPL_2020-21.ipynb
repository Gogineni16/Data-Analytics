```python
#Creating two more columns, Minnutes per match and goals per match
df_epl['Minutespermatch'] = (df_epl['Mins'] / df_epl['Matches']).astype(int)
df_epl['Goalspermatch'] = (df_epl['Goals'] / df_epl['Matches']).astype(float)
```

```python
print(df_epl)
df_epl.head()
```

```python
Total_goals=df_epl['Goals'].sum()
print(Total_goals)
```

```python
#Goals scored in penalities
Total_goals_penalities=df_epl['Penalty_Goals'].sum()
print(Total_goals_penalities)
```

```python
#pie chart for penalties scored vs missed
plt.figure(figsize=(13,6))
plt_not_scored=df_epl['Penalty_Attempted'].sum() - Total_goals_penalities
data= [plt_not_scored, Total_goals_penalities]
labels= [' Penalties missed','Penalties scored']
color=sns.color_palette('Set2')
plt.pie(data, labels= labels, colors= color, autopct='%.0f%%')
plt.show()
```

```python
#Countries with most number of players
nationality=df_epl.groupby('Nationality').size().sort_values(ascending=False)
nationality.head(10).plot(kind= 'bar', figsize=(12,6),color=sns.color_palette("magma"))
```

```python
#clubs with the maximum number of players
df_epl['Club'].value_counts().nlargest(5).plot(kind= 'bar', color=sns.color_palette("viridis"))
```

```python
#nsmallest can be used to plot the clubs with the least number of players 
```

```python
#differentiating players based on age
under20 = df_epl[df_epl['Age']<=20]
age20_25= df_epl[(df_epl['Age']>20) & (df_epl['Age']<=25)]
age25_30= df_epl[(df_epl['Age']>25) & (df_epl['Age']<=30)]
Above30 = df_epl[df_epl['Age']>30]
```

```python
x1= under20['Name'].count()
x2= age20_25['Name'].count()
x3= age25_30['Name'].count()
x4= Above30['Name'].count() 
x= [x1,x2,x3,x4]
mylabels=["<=20",">20 & <=25",">25 & <=30", ">30"]
plt.title('Total players with Age',fontsize=15)
plt.pie(x, labels=mylabels, autopct="%.1f%%")
plt.show()
```

```python
#players under 20 plotted with clubs
players_under20= df_epl[df_epl['Age']<20]
players_under20['Club'].value_counts().plot(kind="bar", color=sns.color_palette("cubehelix"))
```

```python
#AVERAGE age of players in each club
plt.figure(figsize=(12,6))
sns.boxplot(x= 'Club', y= 'Age', data= df_epl)
plt.xticks(rotation=90)
```

```python
num_player=df_epl.groupby('Club').size()
data=(df_epl.groupby('Club')['Age'].sum())/num_player
data.sort_values(ascending=False)
```

```python
#total assists for each club
assists_byclubs= pd.DataFrame(df_epl.groupby('Club', as_index=False)['Assists'].sum())
sns.set_theme(style="whitegrid", color_codes=True)
ax=sns.barplot(x='Club',y='Assists', data=assists_byclubs.sort_values(by="Assists"),palette='Set2')
ax.set_xlabel("Club",fontsize=30)
ax.set_ylabel("Assists",fontsize=20)
plt.xticks(rotation=69)
plt.rcParams["figure.figsize"]=(20,8)
plt.title('Plot of Clubs vs Total Assists', fontsize=20)
```

```python
assists_bygoals= pd.DataFrame(df_epl.groupby('Club', as_index=False)['Goals'].sum())
sns.set_theme(style="whitegrid", color_codes=True)
ax=sns.barplot(x='Club',y='Goals', data=assists_bygoals.sort_values(by="Goals"),palette='rocket')
ax.set_xlabel("Clubs",fontsize=30)
ax.set_ylabel("Goals",fontsize=20)
plt.xticks(rotation=69)
plt.rcParams["figure.figsize"]=(20,8)
plt.title('Plot of Clubs vs Total Goals', fontsize=20)
```

```python
#Top players with highest number of goals
top_10_goals=df_epl[['Name','Club','Goals', 'Matches']].nlargest(n=10, columns='Goals')
top_10_goals
```

```python
#Goals per match
top_10_goals_permatch= df_epl[['Name', 'Goalspermatch','Matches','Goals']].nlargest(n=10, columns='Goalspermatch')
top_10_goals_permatch
```

```python
#plotting goals with assists and without assists
plt.figure(figsize=(14,7))
assists= df_epl['Assists'].sum()
data= [Total_goals-assists, assists]
labels= ['Goals without assists','Goals with assists']
color= sns.color_palette('Set1')
plt.pie(data, labels=labels, colors=color, autopct="%.0f%%")
plt.show()
```

```python
#Players with the most number of yellow cards
epl_yellow=df_epl.sort_values(by='Yellow_Cards',ascending=False)[:10]
plt.figure(figsize=(20,6))
plt.title("Players with the most number of yellow cards")
c=sns.barplot(x=epl_yellow['Name'], y= epl_yellow['Yellow_Cards'], label='Players',color='yellow')
plt.ylabel('Number of yellow cards')
c.set_xticklabels(c.get_xticklabels(),rotation=45)
```
