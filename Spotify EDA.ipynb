```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
tracks_df=pd.read_csv("/Users/avanchowdary/Downloads/archive/tracks.csv")
tracks_df.head()
```

```python
tracks_df.info()
```

```python
pd.isnull(tracks_df).sum()
```

```python
tracks_df.fillna(0, inplace= True)
```

```python
tracks_df.isnull().sum()
```

```python
#Finding the 10 least popular songs
sort_df= tracks_df.sort_values('popularity', ascending= True)
sort_df.head()
```

```python
#Statistics of the data frame.
tracks_df.describe()
#You can also use the transpose function to transpose the whole data frame.
```

```python
#Finding the 15 most popular songs with the popularity more than 90
most_df= tracks_df.query('popularity>90', inplace= False).sort_values('popularity', ascending= False)
most_df.head(15)
```

```python
#Changing the index of the data_frame
tracks_df.set_index('release_date', inplace=True)
tracks_df.index=pd.to_datetime(tracks_df.index)
tracks_df.head()
```

```python
#Finding the artist at a praticular index
tracks_df[["artists"]].iloc[21]
```

```python
tracks_df["duration"]= tracks_df["duration_ms"].apply(lambda x: round(x/1000))
tracks_df.drop("duration_ms", inplace=True, axis=1)
```

```python
tracks_df.duration.head()
```

```python
#Plot a pearson correlation graph
df_corr=tracks_df.drop(["key","mode","explicit"],axis=1).corr(method="pearson")
plt.figure(figsize=(14,6))
heatmap=sns.heatmap(df_corr,annot=True,fmt=".1g",vmin=-1,vmax=1,center=0, cmap="inferno", linewidth=1, linecolor="Black")
heatmap.set_title("correlation heatmap between variable")
heatmap.set_xticklabels(heatmap.get_xticklabels(), rotation=90)
```

```python
#Measuring 0.5% of the dataframe
sample_df=tracks_df.sample(int(0.005*len(tracks_df)))
print(len(sample_df))
```

```python
#Plotting correlation between loudness and energy
plt.figure(figsize=(10,6))
sns.regplot(data=sample_df,y="loudness", x="energy", color="c").set(title="Correlation- Loudness vs Energy")
```

```python
#Plotting correlation between Popularity and Acousticness
plt.figure(figsize=(10,5))
sns.regplot(data=sample_df,y="popularity", x="acousticness", color="b").set(title="Correlation- Popularity vs Acousticness")
```

```python
#Adding a new column using the release date
tracks_df['dates']= tracks_df.index.get_level_values('release_date')
tracks_df.dates=pd.to_datetime(tracks_df.dates)
years=tracks_df.dates.dt.year
```

```python
#EDA of another dataframe - Genre
genre_df= pd.read_csv("/Users/avanchowdary/Downloads/SpotifyFeatures.csv")
```

```python
genre_df.head()
```

```python
#plotting a grpah to depict the duration of songs for different genres
plt.title("Duration of the songs in different genres")
sns.color_palette("rocket",as_cmap=True)
sns.barplot(y="genre", x="duration_ms",data=genre_df)
plt.xlabel("Duration in milli seconds")
plt.ylabel("Genres")
```

```python
#Plotting Top 5 genres by popularity
sns.set_style(style="darkgrid")
plt.figure(figsize= (10,5))
famous=genre_df.sort_values("popularity",ascending=False).head(10)
sns.barplot(y='genre', x='popularity', data=famous).set(title="Top 5 Genres in Popularity")
```
